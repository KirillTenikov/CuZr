## 2. Helper functions for extxyz files

These helpers count structures and create a validation split from the provided training set.

The split logic is:
- keep `test.extxyz` untouched
- split `train.extxyz` into:
  - `train_split.extxyz`
  - `valid_split.extxyz`


```python
def read_extxyz_blocks(path: Path) -> List[List[str]]:
    """Read an extxyz file as a list of structure blocks.

    Assumes standard XYZ/extxyz structure:
    line 1: number of atoms
    line 2: comment / properties
    next N lines: atoms
    """
    path = Path(path)
    blocks = []
    with path.open("r", encoding="utf-8") as f:
        while True:
            first = f.readline()
            if not first:
                break
            first_stripped = first.strip()
            if first_stripped == "":
                continue
            try:
                natoms = int(first_stripped)
            except ValueError as exc:
                raise ValueError(
                    f"Failed to parse atom count in {path}: line starts with {first_stripped!r}"
                ) from exc

            comment = f.readline()
            if comment == "":
                raise ValueError(f"Unexpected EOF after atom-count line in {path}")

            atom_lines = []
            for _ in range(natoms):
                line = f.readline()
                if line == "":
                    raise ValueError(f"Unexpected EOF inside atomic block in {path}")
                atom_lines.append(line)

            blocks.append([first, comment, *atom_lines])
    return blocks


def count_extxyz_structures(path: Path) -> int:
    if not Path(path).exists():
        return 0
    return len(read_extxyz_blocks(path))


def write_extxyz_blocks(path: Path, blocks: List[List[str]]) -> None:
    path = Path(path)
    with path.open("w", encoding="utf-8") as f:
        for block in blocks:
            f.writelines(block)


def make_train_valid_split(
    train_file: Path,
    train_out: Path,
    valid_out: Path,
    valid_fraction: float = 0.10,
    seed: int = 42,
) -> Dict[str, int]:
    blocks = read_extxyz_blocks(train_file)
    n_total = len(blocks)
    if n_total == 0:
        raise ValueError(f"No structures found in {train_file}")

    rng = random.Random(seed)
    indices = list(range(n_total))
    rng.shuffle(indices)

    n_valid = max(1, int(round(n_total * valid_fraction)))
    valid_idx = set(indices[:n_valid])

    train_blocks = [blocks[i] for i in range(n_total) if i not in valid_idx]
    valid_blocks = [blocks[i] for i in range(n_total) if i in valid_idx]

    write_extxyz_blocks(train_out, train_blocks)
    write_extxyz_blocks(valid_out, valid_blocks)

    return {
        "n_total": n_total,
        "n_train": len(train_blocks),
        "n_valid": len(valid_blocks),
        "seed": seed,
    }
```

## 5. Candidate MACE configurations

Instead of doing a large hyperparameter optimization, this notebook uses a **small set of carefully chosen candidates**.

This is the intended strategy for the project:
- fewer runs
- easier comparison
- faster path to scientific MD simulations

The current candidates are designed as sensible production upgrades relative to the DEV model.


```python
cfg_A = {
    "name": "CuZr_MACE_prod_A",
    "r_max": 6.0,
    "num_interactions": 3,
    "hidden_irreps": "128x0e + 128x1o",
    "num_bessel": 10,
    "batch_size": 8,
    "max_num_epochs": 200,
    "lr": 5e-4,
    "weight_decay": 1e-8,
    "scheduler": "ReduceLROnPlateau",
    "ema": True,
}

cfg_B = {
    "name": "CuZr_MACE_prod_B",
    "r_max": 6.0,
    "num_interactions": 4,
    "hidden_irreps": "128x0e + 128x1o",
    "num_bessel": 10,
    "batch_size": 8,
    "max_num_epochs": 250,
    "lr": 5e-4,
    "weight_decay": 1e-8,
    "scheduler": "ReduceLROnPlateau",
    "ema": True,
}

cfg_C = {
    "name": "CuZr_MACE_prod_C",
    "r_max": 7.6,
    "num_interactions": 3,
    "hidden_irreps": "128x0e + 128x1o",
    "num_bessel": 12,
    "batch_size": 6,
    "max_num_epochs": 250,
    "lr": 3e-4,
    "weight_decay": 1e-8,
    "scheduler": "ReduceLROnPlateau",
    "ema": True,
}

cfg_D = {
    "name": "CuZr_MACE_prod_D",
    "r_max": 6.0,
    "num_interactions": 3,
    "hidden_irreps": "64x0e + 64x1o",
    "num_bessel": 10,
    "batch_size": 10,
    "max_num_epochs": 200,
    "lr": 5e-4,
    "weight_decay": 1e-8,
    "scheduler": "ReduceLROnPlateau",
    "ema": True,
}

configs = [cfg_A, cfg_B, cfg_C, cfg_D]
pd.DataFrame(configs)
```

### Why these candidates?

- **A**: conservative main production candidate  
- **B**: deeper model  
- **C**: larger cutoff radius  
- **D**: smaller/faster comparison model  

A good practical workflow is:
1. train all candidates,
2. compare validation metrics,
3. check LAMMPS export,
4. check Notebook 05/06 physics tests,
5. choose the final production model.


## 6. Command generation for MACE training

The generated commands are intended for **A100 or B200** runs.

### A100 vs B200
- If the B200 environment is stable and CUDA/PyTorch are already working there, it is a strong choice.
- If the B200 software stack is uncertain, A100 is the safer fallback.

The command below uses `--device=cuda`, which is correct for either A100 or B200.


```python
def make_mace_command(cfg: Dict, train_file: Path, valid_file: Path, test_file: Path | None = None, device: str = "cuda") -> str:
    cmd = [
        "mace_run_train",
        f"--name={cfg['name']}",
        f"--train_file={train_file}",
        f"--valid_file={valid_file}",
        f"--r_max={cfg['r_max']}",
        f"--num_interactions={cfg['num_interactions']}",
        f"--hidden_irreps={cfg['hidden_irreps']!r}",
        f"--num_bessel={cfg['num_bessel']}",
        f"--batch_size={cfg['batch_size']}",
        f"--max_num_epochs={cfg['max_num_epochs']}",
        f"--lr={cfg['lr']}",
        f"--weight_decay={cfg['weight_decay']}",
        f"--scheduler={cfg['scheduler']}",
        f"--device={device}",
    ]
    if cfg.get("ema", False):
        cmd.append("--ema")
    if test_file is not None:
        cmd.append(f"--test_file={test_file}")
    return " \\
  ".join(cmd)


for cfg in configs:
    print("=" * 100)
    print(cfg["name"])
    print(make_mace_command(cfg, TRAIN_SPLIT_FILE, VALID_SPLIT_FILE, TEST_FILE, DEVICE))
    print()
```

## 7. Save commands and configs to disk

This makes the training preparation reproducible and easy to transfer to the GPU machine.


```python
commands_txt = OUTPUT_DIR / "mace_candidate_commands.txt"
configs_json = OUTPUT_DIR / "mace_candidate_configs.json"

with commands_txt.open("w", encoding="utf-8") as f:
    for cfg in configs:
        f.write("=" * 100 + "\n")
        f.write(cfg["name"] + "\n")
        f.write(make_mace_command(cfg, TRAIN_SPLIT_FILE, VALID_SPLIT_FILE, TEST_FILE, DEVICE) + "\n\n")

with configs_json.open("w", encoding="utf-8") as f:
    json.dump(configs, f, indent=2)

print("Saved:", commands_txt)
print("Saved:", configs_json)
```

## 9. Model selection criteria

The final production MACE model should be chosen based on **both ML metrics and physical behavior**.

### ML-side criteria
- training stability
- validation loss
- energy/force error
- no pathological overfitting

### Physics-side criteria
- stable export to LAMMPS
- smooth crystalline EOS
- reasonable smoke MD behavior
- reasonable amorphous structure and RDF
- later: NCL-relevant behavior in production MD

This is important: the best model is not necessarily the one with the very lowest validation loss if its physical behavior is worse.
